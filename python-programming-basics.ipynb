### switch Statement (match keyword)

```python
#it will work in python 3.10 and above
colour = "yellow"
match colour:
    case "white":
        print("The specified colour is white")
    case "green":
        print("Wow, you've picked green")
    case "blue":
        print("Great, you are going with blue colour")
    case "yellow":
        print("Yellow like mango...")
    case other:
        print("No match found!")
```

## 2.6 Loops

### while loop

```python
# print the value of i until it is less than 6
i = 1
while i < 6:
    print(i)
    i+=1
```

```python
# the break statement is used when we want the while
# loop to stop even if the condition is true. See the 
# following example, where we want to stop the while
# loop even if the i is less than 6. As soon as the
# loop reaches i==3, it breaks and ignores all upcoming 
# statements

i = 0
while i<6:
    i+=1
    if(i==3):
        break
    print(i)
```

```python
# By using the continue statement, we can stop the
# current iteration and continue with the next. Considering
# the following example, we ignore the iteration when i==3,
# and continue with the later iterations.

i = 0
while i<6:
    i+=1
    if(i==3):
        continue
    print(i)
```

### for loop

```python
# here we are also iterating to print the value of i
# ranging from 0 to 6 using for loop (6 is not included here)
for i in range(6):
    print(i)
```

```python
# the range(start, stop, step) values are 
# range (1,6,2) to start printing from 1,
# increment by 2 each time. Remember, the
# endpoint is excluded

for i in range(1,6,2):
    print(i)
```
