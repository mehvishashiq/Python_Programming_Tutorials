In this notebook, we are going to learn Python basics that are used in almost every programming language but the syntax of using them differs. Let's start with learning the `print` statement.

## 2.1 `print` Statement

We use a `print()` method to print something on the console. Anything inside this function will be printed on the screen, it can be a `string` type value or a value of a `variable` (we will learn what the `variable` is in a while). Look at the simplest example below.

```python
print("Hi! Let's Learn Python!") # wrote a string value within double-qoutes
print('Hi! Let\'s Learn Python!') # wrote a string value withing single-qoutes
```

Both `print` statements are printing the same message, first statement is using `double-quotes` while second is using `single-qoutes`, which means we can use both, single-quotes and double-quotes to print a `string` but where to use which quotes?

If we know that we'll not have single-quotes within our message then we can use single-quotes in `print()` method; otherwise using double-quotes is preffered because it will save us from using `\'` to print `'` in our message.

**NOTE:** In the above example code, the `#` symbol is used for inline comments while `"""` (three quotes) are used for multi-line comments. The comments are ignored by the compiler & interpreters, they are just for the software developers to make the code easy to manage and understandable.

This is used to interact with the user either to display something or to get more information. In Python, we are using `print()` method to display something on the screen.

## 2.4 Variable

A `variable` is  container that occupies a memory space to store information or data. In Python, we don’t need to explicitly declare a variable of a specific data type but the variable decides its data type depending on what kind of data it is containing.

Note that, the memory addresses are in hexadecimal values while variables are just the names to those hexadecimal values, assigned to those particular memory location to access it easily.See the following example to have a clear understanding.

```python
first_name = "Mehvish"
last_name = 'Ashiq'
print(first_name + " " + last_name) # we can get the same output using as print(first_name, last_name)
print(f'FirstName: {first_name} LastName: {last_name}')
```

In the above code example, `first_name` and `last_name` are two variables that hold string type values. Then, we use `print` statements to print the values of `first_name` and `last_name` variables.

We can print in different ways, the first `print` statement prints the values separated with a single space. To have that single space, we use `+` to concatenate. Now, think of 10, 20, or 30 values, can we use that much `+` to concatenate? Of course not!

This is where we use `f-string` to format easily, we `f` at the start, write whole string within `single-quotes` and variables within the curly braces. We can see sample use in the second `print` statement above.

**Note:** We don’t need semi-colon (;) to end the statement as we do in other programming languages, we only need white space to terminate each statement. Code indentation is very important specifically when we are coding in Python.

### Numbers

As the name suggests, the `Number` data type contains numeric data. It further has three types `int`, `float`, and `complex`. Here, `int` means integer, which contains the whole positive and negative numbers of unlimited length. The `float` type number can hold a floating-point number. It can be positive or negative numbers having one more decimal.

The `complex` type of numbers is those where we have an imaginary part. These types of numbers are in `a+bj` form where `a` and `b` are floating-point numbers and `j` is the square root of `-1`. We will learn a basic example of `complex` numbers but I have not used them in projects.

```python
int_number = 30
float_number = 30.05
complex_number = 5+6j

print(int_number, float_number, complex_number)
```

We can use the `type()` function to confirm their data types as follows:

```python
int_number = 30
float_number = 30.05
complex_number = 5+6j

print(type(int_number), type(float_number), type(complex_number))
```

Let's play with `Number` data type using arithmetic operators.

```python
print(3+5) # we are adding two numbers using plus operator
print(5-3) # we subtract one number from the other using subtraction operator
print(5*3) # we multiply one number with the other using multiplication operator
print(10/3) # we perform division using divide operator
print(5%3) # we use modulus operator to get a remainder
print(2**2) # we use exponentiation operator to perform 2 raise to power 2 (2^2)
print(5//3) # Here, we use floor division, which rounds the result down to a nearest whole number
```

### String

The string data type is very popular in Python and is used to store textual data. It is either enclosed by single quotes or double quotes. The `'hi'` and `"hi"` are the same for Python. Python does not support the `char` data type. We can store one character that acts like a string. Let’s practice it with some examples given below.

```python
character = 'c'
print(character, type(character))
```

```python
first_name = "Mehvish"
print("Hi, I'm: " + first_name)
print('Hi, I\'m: ' + first_name)
```

Did you understand where to use single quotes or double quotes? My suggestion is to use double quotes to enclose the string when you have single quotes within the string because it is easier to manage (see second `print` statement in the above example).

**Note:** The variable name must be descriptive. If we have multiple words in a variable name then each word would be separated by an underscore as we are separating `first_name`.

We enclose the multiline string within the three double quotes (`"""`) to display on multiple lines. See the following example.

```python
message = """Hi, are you learning Python?
Can you help me because I'm also learning it?
"""

print(message)
```

```python
message = "Hello World"
print(len(message)) # len() returns the length of string.
print(message[0]) # get character at index 0
print(message[1:3]) # get characters starting from index 1 (including) to index 3 (excluding)
print(message[:3]) # get characters starting from index 0 to 3 (where index 3 is not included)
print(message[3:]) # get characters starting from index 3 till the last index
```

Let's learn few built-in methods to manipulate string values.

```python
message = "Hello"
print(message.count('l')) # returns the count of the given character
print(message.lower()) # transforms all characters to lower-case
print(message.upper()) # transforms all characters to upper-case
print(message.find('ell')) # returns the starting index of the specified word
```

```python
full_name = "Tahir Raza"
print(full_name)

full_name = full_name[:6]  + "Ashiq" # update string by accessing the index
print(full_name)

full_name = full_name.replace("Ashiq", "Raza") # udpate string using built-in method
print(full_name)
```

Let's learn few special operators in string.

```python
message = "Hello"
print(message + "Python") # we concatenate the value of message variable with "Python"
print(message*3) # we print the value of message thrice
print(message[1]) # we retrieve and print the value at index 1
print(message[2:3]) # we retrieve and print the value starting form index 2 to 3 where index 3 is excluded
print("H" in message) # check if "H" is in message, if it is so, we'll get true otherwise false
print("l" not in message) # check if "l" is not in message, if it is so, we'll get true otherwise false
print("I'm %s and learning %s" % ("Mehvish", "Python")) # use formatting operator to concatenate various types of data
```

We are using various operators that are very useful while working with the string. For instance, `+` concatenates the string, `*` make the `x` number of copies (we are making 3 copies in the above example code).

Passing a number in square brackets (`[]`) ) returns the character at that index while putting a colon in between brackets `[:]` return a range of characters (see the above example).

The `in` operator outputs `True` if the specified character/string is found while the `not in` operator returns `True` when it can’t find the given character/string. The formatting operator `%` is very useful when we want to print various types of data in one statement.
